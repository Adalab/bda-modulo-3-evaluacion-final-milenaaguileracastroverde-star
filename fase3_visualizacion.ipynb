```python
import pandas as pd #hago las importaciones de librerias.
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import warnings # para que no se imprimam los future warning
warnings.filterwarnings('ignore', category=FutureWarning)
```

```python
pd.set_option('display.max_columns', None) 
```

```python
df_limpio = pd.read_csv("Files/df_limpio.csv")
```

1. ¿Cómo se distribuye la cantidad de vuelos reservados por mes durante el año?

```python
sns.barplot(data=df_limpio, x='month', y='total_flights',errorbar=None)
plt.title('Total de vuelos por mes')
```

Claramente los clientes vuelan mas en los periodos vacacionales donde aumenta la cantidad de vuelos.


Existe una relación entre la distancia de los vuelos y los puntos acumulados por los cliente

```python
sns.scatterplot(data=df_limpio, x='points_accumulated', y='distance')
plt.title('Relación entre puntos acumulados y distancia recorrida')
```

A mayor distancia mayor cantidad de puntos acumulados

 ¿Cuál es la distribución de los clientes por provincia o estado?

```python
sns.countplot(data=df_limpio, y='province', palette='viridis', order=df_limpio['province'].value_counts().index)
plt.title('Distribución de clientes por provincia')
```

La cantidad de clientes es proporcional a la densidad demografica de la provincia

¿Cómo se compara el salario promedio entre los diferentes niveles educativos de los clientes?

```python
sns.barplot(x='education', y='salary', palette='viridis', data=df_limpio, errorbar=None)
plt.title('CLV promedio por nivel educativo')
```

Los que tienen un doctorado tienen un salario mayor que los de nivel educativo inferior

¿Cuál es la proporción de clientes con diferentes tipos de tarjetas de fidelidad?

```python
sns.countplot(data=df_limpio, y='loyalty_card', palette='viridis', order=df_limpio['loyalty_card'].value_counts().index)
plt.title('Distribución de clientes por tipo de tarjeta de fidelidad')
```

Los clientes con mayores recursos y formacion son los que mas gastan, y por lo tanto los que mas valor tienen para la empresa. Por otro lado, los clientes con tarjeta de fidelidad "Star" son los que mas vuelos han reservado, lo que indica que son clientes frecuentes y valiosos para la empresa.

¿Cómo se distribuyen los clientes según su estado civil y género?

```python
sns.countplot(data=df_limpio, x= 'marital_status', hue ='gender', palette='viridis')
plt.title('Distribución de clientes por estado civil y género')
```

La mayoria de los clientes son casados son distincion de genero.
