```python
df_1.isna().sum()/df_1.shape[0]*100 # busco el porcentaje de valores nulos por columna, divido el numero de valores nulos por el numero total de filas y multiplico por 100 para obtener el porcentaje
```

```python
df = pd.merge(df_1, df_2, on='Loyalty Number', how='left') # hago un merge de los dos dataframes utilizando la columna "Loyalty Number" como clave, y utilizando un left join para que se mantengan todas las filas del dataframe 1 y se añadan las columnas del dataframe 2 donde coincidan los valores de "Loyalty Number". Las filas del dataframe 1 que no tengan un valor coincidente en el dataframe 2 tendrán valores nulos en las columnas añadidas.
```

```python
df.head(5) # compruebo que el merge se ha hecho correctamente y que se han añadido las columnas del dataframe 2 al dataframe 1.
```

```python
df.columns= df.columns.str.lower().str. replace(" ", "_") # elimino los espacios en blanco de los nombres de las columnas para evitar problemas a la hora de acceder a ellas.
```

```python
df.drop(columns=["country"], inplace=True) # elimino la columna "Country" porque tiene un valor unico con una alta frecuencia y no aporta información relevante.
```
