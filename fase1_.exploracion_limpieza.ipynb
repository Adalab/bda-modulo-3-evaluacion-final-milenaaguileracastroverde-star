```python
df_1.isna().sum()/df_1.shape[0]*100 # busco el porcentaje de valores nulos por columna, divido el numero de valores nulos por el numero total de filas y multiplico por 100 para obtener el porcentaje
```

```python
df = pd.merge(df_1, df_2, on='Loyalty Number', how='left') # hago un merge de los dos dataframes utilizando la columna "Loyalty Number" como clave, y utilizando un left join para que se mantengan todas las filas del dataframe 1 y se añadan las columnas del dataframe 2 donde coincidan los valores de "Loyalty Number". Las filas del dataframe 1 que no tengan un valor coincidente en el dataframe 2 tendrán valores nulos en las columnas añadidas.
```

```python
df.head(5) # compruebo que el merge se ha hecho correctamente y que se han añadido las columnas del dataframe 2 al dataframe 1.
```

```python
df.columns= df.columns.str.lower().str. replace(" ", "_") # elimino los espacios en blanco de los nombres de las columnas para evitar problemas a la hora de acceder a ellas.
```

```python
df.drop(columns=["country"], inplace=True) # elimino la columna "Country" porque tiene un valor unico con una alta frecuencia y no aporta información relevante.
```

```python
df.columns #compruebo que las modificaciones y la eliminacion de columna se han hecho correctamente.
```

```python
df.to_csv('Files/df_nuevo.csv', index=False) # guardo el nuevo dataframe en un csv para poder utilizarlo en el futuro sin tener que hacer el merge de nuevo.
print("DataFrame guardado como 'df_nuevo.csv'")
```

Interpretación de la Fase 1 - Exploración y Limpieza

Tras completar la fase de preparación de datos, se han realizado las siguientes acciones críticas para garantizar un análisis estadístico fiable:

- Unificación de Datos: Se han combinado los datasets de actividad de vuelo (Customer Flight Analysis.csv) y el historial de lealtad (Customer Loyalty History.csv) mediante un left join sobre la columna loyalty_number. Esto permite cruzar el comportamiento de vuelo con el perfil demográfico de cada cliente.

- Estandarización: Se han transformado todos los nombres de las columnas a minúsculas y se han sustituido los espacios por guiones bajos para mejorar la legibilidad y evitar errores de sintaxis en el código.

- Tratamiento de Inconsistencias: Se detectaron valores negativos en la columna salary, los cuales fueron corregidos mediante su valor absoluto .(abs) al identificarse como errores de entrada de datos.

- Gestión de Valores Nulos: En la variable salary, se optó por la imputación mediante la mediana para no perder el volumen de datos de esos clientes, evitando así sesgos por valores extremos. En las columnas de Cancellation Year y Cancellation Month, los valores nulos se mantuvieron entendiendo que representan a clientes que aún mantienen su membresía activa.

- Integridad: Se verificó la ausencia de registros duplicados para asegurar que cada transacción y perfil sea único, cumpliendo con los estándares de consistencia requeridos.

Conclusión: El conjunto de datos resultante está optimizado y listo para el análisis estadístico y la visualización de las siguientes fases.
